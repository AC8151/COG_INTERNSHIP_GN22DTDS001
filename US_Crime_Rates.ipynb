##Step 8. Group the year by decades and sum the values
Pay attention to the Population column number, summing this column is a mistake

```python
type(crime.index)
```

```python
## resampling this time-series data based on groups of 10 years (DECADES)

# 10AS -> 10 years (A), from starting date i.e. 1960-01-01 (S)
crimes_decade = crime.resample('10AS').sum()

# Use resample to get the maximum value of the "Population" column, 
#i.e., the population at the end of every decade serves as the population for that decade
crimes_decade['Population'] = crime['Population'].resample('10AS').max()
crimes_decade
```

##Step 9. What is the most dangerous decade to live in the US?

```python
crimes_decade.idxmax(axis=0)
```

*OBSERVATION: Based on this, we can conclude that the 90's was the most dangerous decade to live in the US.*
